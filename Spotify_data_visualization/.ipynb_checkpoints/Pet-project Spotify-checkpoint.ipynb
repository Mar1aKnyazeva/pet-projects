## Import libraries | Импорт библиотек

```python
import pandas as pd # linear algebra | линейная алгебра
import numpy as np # data processing, csv file | обработка данных, работа с csv форматом файлов
import matplotlib.pyplot as plt # plotting and visualisation | построение графиков и визуализация
import matplotlib.ticker as mtick # custom tick marks and labels for plots | настройка отметок и меток на графиках
import matplotlib.gridspec as gridspec # grid layout for subplots | сетка для расположения подграфиков
import seaborn as sns # statistical data visualization, based on Matplotlib | статистическая визуализация данных на основе Matplotlib
```

## Loading dataset and initial review | Загрузка датасета и первичный обзор

```python
# file location on GitHub | расположение файла на GitHub
url = 'https://github.com/Mar1aKnyazeva/pet-projects/raw/main/Spotify_data_visualization/Most%20Streamed%20Spotify%20Songs%202024.csv'

# reading .csv-file, using 'ISO-8859-1' - an encoding designed for Western European languages
# Чтение csv-файла, используя 'ISO-8859-1' - кодировка для западноевропейских языков
df = pd.read_csv(url, encoding='ISO-8859-1')
```

```python
# examining the first five lines to get an idea of the structure and content
# Осмотр первых пяти строк, чтобы получить представление о структуре и содержимом  
print(display(df.head()))
```

```python
# Getting information about data types, the number of non-zero values and the size of the dataset
# Получение информацию о типах данных, количестве ненулевых значений и размере датасета
print(df.info())
```

```python
# Getting basic statistics on numeric columns: mean, standard deviation, minimum and maximum values
# Получение базовой статистики числовых столбцов: среднее, стандартное отклонение, минимальные и максимальные значения
print(display(df.describe()))
```

## Checking and cleaning data | Проверка и очистка данных

```python
# Removing duplicate rows, based on the 'Track' and 'Artist' columns, leaving one entry with the maximum number of 'Spotify Streams'
# Удаление повторяющихся строк на основе столбцов 'Трек' и 'Исполнитель', оставляя одну запись с максимальным количеством 'Spotify Streams'
df_cleaned = df.sort_values('Spotify Streams', ascending=False).drop_duplicates(subset=['Track', 'Artist'])

# Removing empty rows in the "Spotify Streams' column
# Удаление пустых строк в столбце 'Spotify Streams'
df_cleaned = df_cleaned.dropna(subset=['Spotify Streams'])
```

```python
# Convert 'Release Date' column to datetime format
# Преобразование столбца 'Release Date' в формат datetime
df_cleaned['Release Date'] = pd.to_datetime(df_cleaned['Release Date'], format ='%m/%d/%Y')
```

```python
# Convert 'Spotify Streams' column to number format
# Преобразование столбца 'Spotify Streams' в числовой формат
df_cleaned['Spotify Streams'] = df_cleaned['Spotify Streams'].str.replace(',', '').astype(int)
```

## Release time analysis | Анализ по времени релиза

```python
# Extrack "year" and "month" into separate new columns
# Извлечение "года" и "месяца" в отдельные новые столбцы
df_cleaned['Release Year'] = df_cleaned['Release Date'].dt.year
df_cleaned['Release Month'] = df_cleaned['Release Date'].dt.month

# Creating a figure
# Создание фигуры
plt.figure(figsize=(12,8))
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette(palette='Greens_d')

# Analysis of the number of released tracks by year
# Анализ количества выпущенных треков по годам
ax1 = plt.subplot(2,1,1) 
releases_per_year = df_cleaned['Release Year'].value_counts().sort_index()
releases_per_year.plot(kind='bar')
ax1.set_title('Number of releases by year', fontsize=14, weight='bold')
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of releases', fontsize=12)

# Analysis of track popularity by year
# Анализ популярности треков по годам
ax2 = plt.subplot(2,1,2)
popularity_per_year = df_cleaned.groupby('Release Year')['Track Score'].mean()
popularity_per_year.plot(kind='line', marker='s')
ax2.set_title('Average Popularity of tracks by year', fontsize=14, weight='bold')
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Average Track Score', fontsize=12)

# Display of graphs
# Отображение графиков
plt.tight_layout()
plt.show()
```

```python
# Creating a figure
# Создание фигуры
plt.figure(figsize=(12,8))
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette(palette='Greens_d')

# Analysis of the number of released tracks by month
# Анализ количества выпущенных треков по месяцам
ax1 = plt.subplot(2,1,1)
releases_per_month = df_cleaned['Release Month'].value_counts().sort_index() 
releases_per_month.plot(kind='bar')
ax1.set_title('Number of releases by month', fontsize=14, weight='bold')
plt.xlabel('Release Month', fontsize=12)
plt.xticks(ticks=np.arange(len(releases_per_month)), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], fontsize=12, rotation=0)
plt.ylabel('Number of releases', fontsize=12)

# Analysis of track popularity by year
# Анализ популярности треков по годам
ax2 = plt.subplot(2,1,2)
popularity_per_month = df_cleaned.groupby('Release Month')['Track Score'].mean()
popularity_per_month.plot(kind='line', marker='s')
ax2.set_title('Average Popularity of tracks by month', fontsize=14, weight='bold')
plt.xlabel('Release Month', fontsize=12)
plt.xticks(ticks=np.arange(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], fontsize=12, rotation=0)
plt.ylabel('Average Track Score', fontsize=12)

# Display of graphs
# Отображение графиков
plt.tight_layout()
plt.show()
```

## Track popularity analysis | Анализ популярности треков

```python
# Top-10 Most streamed tracks, based on 'Spotify Streams'
# Топ-10 самых прослушиваемых треков, основанный на 'Spotify Streams'

# Data sampling and processing
# Выборка и обработка данных
top_tracks = df_cleaned[['Track', 'Artist', 'Spotify Streams']].sort_values(by='Spotify Streams', ascending=False).head(10)
top_tracks['Position'] = range(1, 11)
top_tracks['Track & Artist'] = top_tracks['Track'] + ' (' + top_tracks['Artist'] + ')' 
top_tracks['Streams'] = top_tracks['Spotify Streams'].apply(lambda x: f'{x/1e6:.2f}M')

# Creating a figure
# Создание фигуры
plt.figure(figsize=(14, 8))
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette(palette='Greens_d')
plt.suptitle('Top 10 Tracks by Spotify Streams for 2024', fontsize=16, weight='bold')

# Creating a table
# Создание таблицы
ax0 = plt.subplot(2, 1, 1)
ax0.axis('off')
col_widths = [0.15, 0.6, 0.35, 0.25]
table = ax0.table(
    cellText=top_tracks[['Position', 'Track', 'Artist', 'Streams']].values,
    colLabels=['Position', 'Track', 'Artist', 'Streams'],
    cellLoc='center',
    loc='center',
    colWidths=col_widths
)
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 1.6)

# Table styling
# Стилизация таблицы
for key, cell in table.get_celld().items():
    cell.set_edgecolor('Grey')
    if key[0] == 0: #если это заголовок
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#9EC99E')
    else:
        cell.set_facecolor('#D8F0D8' if key[0] % 2 == 0 else 'white')

# Creating a graph
# Создание графика
ax1 = plt.subplot(2, 1, 2)
sns.barplot(x='Spotify Streams', y='Track & Artist', hue='Track & Artist', legend=False, data=top_tracks, palette='Greens_d', errorbar=None)
ax1.set_xlabel('Number of streams', fontsize=14)
ax1.set_ylabel('Track', fontsize=14)
ax1.set_yticks(range(len(top_tracks)))
ax1.set_yticklabels(top_tracks['Track & Artist'], fontsize=12)
formatter = mtick.FuncFormatter(lambda x, _: f'{x/1e6:.2f}M')
ax1.xaxis.set_major_formatter(formatter)

# Display of table and graphs
# Отображение таблицы и графика
plt.subplots_adjust(hspace=0.2) 
plt.tight_layout(rect=[0, 0, 1, 0.99]) 
plt.show()
```

## Artists popularity analysis | Анализ популярности исполнителей

```python
# Top-10 most popular artists, based on 'Spotify Streams'
# Топ-10 самых популярных исполнителей, основанный на Spotify Streams

# Data sampling and processing
# Выборка и обработка данных
top_artist_streams = df_cleaned.groupby('Artist')['Spotify Streams'].sum().sort_values(ascending=False).head(10)
top_artist_tracks = df_cleaned.groupby('Artist')['Track'].count()
top_artist = pd.DataFrame({
    'Position': range(1, (len(top_artist_streams)+1)),
    'Artist': top_artist_streams.index,
    'Spotify Streams': top_artist_streams.values,
    'Number of Tracks': top_artist_tracks[top_artist_streams.index].values  
})
top_artist['Streams'] = top_artist['Spotify Streams'].apply(lambda x: f'{x/1e6:.2f}M')

# Creating a figure
# Создание фигуры
plt.figure(figsize=(14, 8))
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette(palette='Greens_d')
plt.suptitle('Top 10 Artists by Spotify Streams for 2024', fontsize=16, weight='bold')

# Creating a table
# Создание таблицы
ax0 = plt.subplot(2, 1, 1)
ax0.axis('off')
col_widths = [0.15, 0.6, 0.35, 0.25]
table = ax0.table(
    cellText=top_artist[['Position', 'Artist', 'Streams', 'Number of Tracks']].values,
    colLabels=['Position', 'Artist', 'Streams', 'Number of Tracks'],
    cellLoc='center',
    loc='center',
    colWidths=col_widths
)
table.auto_set_font_size(False) 
table.set_fontsize(12) 
table.scale(1.1, 1.6) 

# Table styling
# Стилизация таблицы
for key, cell in table.get_celld().items():
    cell.set_edgecolor('Grey')
    if key[0] == 0: #если это заголовок
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#9EC99E')
    else:
        cell.set_facecolor('#D8F0D8' if key[0] % 2 == 0 else 'white')

# Creating a graph
# Создание графика
ax1 = plt.subplot(2, 1, 2)
sns.barplot(x='Spotify Streams', y='Artist', hue='Artist', data=top_artist, palette='Greens_d', legend=False, errorbar=None)
ax1.set_xlabel('Number of streams', fontsize=14)
ax1.set_ylabel('Artist', fontsize=14)
ax1.set_yticks(range(len(top_artist)))
ax1.set_yticklabels(top_artist['Artist'], fontsize=12)
formatter = mtick.FuncFormatter(lambda x, _: f'{x/1e6:.2f}M')
ax1.xaxis.set_major_formatter(formatter)

# Display of table and graphs
# Отображение таблицы и графика
plt.subplots_adjust(hspace=0.2)
plt.tight_layout(rect=[0, 0, 1, 0.99])
plt.show()
```
